```python
decode = data.get("groups")[0]
```

```python
decode_otravez = decode.get("items")
```

```python
mapa_nombre =  ["venue", "name"]
mapa_longitud = ["venue", "location", "lng"]
mapa_latitud = ["venue", "location", "lat"]
```

```python
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
companies = []
for dic in decode_otravez:
    paralista = {}
    paralista["name"] = getFromDict(dic, mapa_nombre)
    paralista["longitude"]= getFromDict(dic, mapa_longitud)
    paralista["latitude"] = getFromDict(dic,mapa_latitud)
    companies.append(paralista)
```

```python
companies_data = pd.DataFrame(companies)
```

```python
companies_data
```

```python
companies_data.dropna(inplace=True)
```

```python
companies_data['location']= ''
```

```python
for index, row in companies_data.iterrows():
    companies_data["location"][index] = Point((row['longitude'], row['latitude']))
```

```python
companies_data = gpd.GeoDataFrame(companies_data, geometry=gpd.points_from_xy(companies_data['longitude'], companies_data['latitude']))
```

```python
Map([Layer(companies_data, 'color: red', encode_data=False,popup_hover = [popup_element("name", "Companies" )])])
```
